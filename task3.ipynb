```python
client_df = pd.read_csv('client_data (1).csv')
client_df["date_end"]= pd.to_datetime(client_df["date_end"],format='%Y-%m-%d')
client_df.head()
```

```python
# Ensure date_activ and date_end are in datetime format
client_df["date_activ"] = pd.to_datetime(client_df["date_activ"])
client_df["date_end"] = pd.to_datetime(client_df["date_end"])

# Now perform the subtraction
client_df["customer_tenure"] = (client_df["date_end"] - client_df["date_activ"]).dt.days

# Display the first few values to verify
print(client_df["customer_tenure"].head())
```

```python
#Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

```python
# Group prices by ID and quarter
quarterly_price_by_id = price_df.groupby(['id', pd.Grouper(key='price_date', freq='Q')]).agg({
    'price_off_peak_var': 'mean',
    'price_off_peak_fix': 'mean'
}).reset_index()

# Get Q1 (Jan-Mar) and Q4 (Oct-Dec) prices
q1_prices = quarterly_price_by_id.groupby('id').first().reset_index()
q4_prices = quarterly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
q_diff = pd.merge(q4_prices.rename(columns={'price_off_peak_var': 'q4_var', 'price_off_peak_fix': 'q4_fix'}),
                  q1_prices.drop(columns='price_date'),
                  on='id')
q_diff['offpeak_diff_q4_q1_energy'] = q_diff['q4_var'] - q_diff['price_off_peak_var']
q_diff['offpeak_diff_q4_q1_power'] = q_diff['q4_fix'] - q_diff['price_off_peak_fix']
q_diff = q_diff[['id', 'offpeak_diff_q4_q1_energy', 'offpeak_diff_q4_q1_power']]
q_diff.head()
```

```python
# Sort data by ID and price_date
price_df = price_df.sort_values(['id', 'price_date'])

# Calculate month-over-month price change
price_df['offpeak_var_monthly_change'] = price_df.groupby('id')['price_off_peak_var'].diff()
price_df['offpeak_fix_monthly_change'] = price_df.groupby('id')['price_off_peak_fix'].diff()

# Show results
price_df[['id', 'price_date', 'offpeak_var_monthly_change', 'offpeak_fix_monthly_change']].head()
```

```python
# Calculate the ratio between peak and off-peak prices
price_df['peak_offpeak_ratio_var'] = price_df['price_peak_var'] / price_df['price_off_peak_var']
price_df['peak_offpeak_ratio_fix'] = price_df['price_peak_fix'] / price_df['price_off_peak_fix']

# Show results
price_df[['id', 'price_date', 'peak_offpeak_ratio_var', 'peak_offpeak_ratio_fix']].head()
```

```python
# Calculate rolling mean (trend) over 3 months
price_df['rolling_mean_3m_var'] = price_df.groupby('id')['price_off_peak_var'].rolling(window=3).mean().reset_index(level=0, drop=True)
price_df['rolling_mean_3m_fix'] = price_df.groupby('id')['price_off_peak_fix'].rolling(window=3).mean().reset_index(level=0, drop=True)

# Show results
price_df[['id', 'price_date', 'rolling_mean_3m_var', 'rolling_mean_3m_fix']].head()
```
